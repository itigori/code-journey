# 课程表

## 说明

基于ComponentV2的课程表界面，基本实现课程表全功能，且具有高自定义性的UI

## 调用示例
``` typescript
// 课程信息
private scheduleList: CourseInfo[] = [
    new CourseInfo(
        [1, 2],           // 课程占用节数
        1,                // 课程所在星期几
        [1, 2, 3, 4, 5],  // 课程所在周
        '鸿蒙系统开发',    // 课程名字
        '#0080FF',      // 课程卡片边框色
        '#66CCFF',      // 课程卡片主色
        'Chino Yuki',     // 课程老师
        'School Campus',  // 课程校区
        'Class Location', // 课程地点
        'Class Type'      // 课程类型
    )
]


// 课程表
@State today: new Date()
@State semesterStart: Date = new Date(2026, 1, 6)


build() {
    Column() {
        ScheduleFlow({
            currentDate: this.today,               // 今天
            semesterStartDate: this.semesterStar,  // 学期的第一天
            totalWeekCount: 20,                    // 课程表总周数
            divideTimes: [                         // 课程上下午节数
                new TimeInfo(4, 'AM'),
                new TimeInfo(9, 'PM')
            ],
            scheduleMaxIndex: 9,                   // 课程表总节数
            scheduleList: this.scheduleList,       // 课程数组
            scheduleClick: () => {}                // 点击事件
        })
    }
}
```

# BreakPoint - 断点系统
用于在任何component模块中监测屏幕实时尺寸，来对UI/UX做响应式反馈。

相关文件：
EntryAbility.ets: 在打开应用时注册断点
BreakPointType.ets: 获取断点值

编写版本：API 20

## 使用范例

``` TypeScript
import { BreakPointType } from BreakPointType

@Component
export struct sample {
      @StorageProp('currentBreakPoint') bp: string = '' // 监听断点

      build {
        Column {
            text('Hello Takanashi')
            .fontSize(new BreakPointType({
                sm: 12,
                md: 14,
                lg: 16
            }).getValue(this.bp))
        }
      }
}
```

# Python 字典
字典用于存储键和值，键用来查找值
```python
contact = {
    "A" = "1234567"
    "B" = "7654321"
}

# 获取键的值
contact["A"]

# 元组 tuple 元组可以放多个元素
name_tuple = ("A", "32") # 不可变

contact = {
    ("A", "32") = "1234567"
}

# 添加值 如果键已存在，则元素会覆盖
contact["C"] = "999"
# 删除值 如果键不存在，则会报错
del contact["A"]
# 字典长度
len(contact)

contact.keys() # 返回所有键
contact.values() # 返回所有值
contact.items() # 返回所有键值对
```

# Python 类

``` python
# 父类
class Mammal:
    def __init__(self, name, sex):
        self.name = name
        self.sex = sex
    
    def breathe(self):
        print(self.name + "Breathing")

    def poop(self):
        print(self.name + "Pooping")

# 子类
class Human(Mammal):
    def read(self):
        print(self.name + "Reading")

class Cat(Mammal):
    def scratch(self):
        print(self.name + "Scratching")
```

super() 子继承父元素

``` python
class Father:
    def __init__(self, name):
        self.name = name

class Son:
    def __init__(self, name):
        super().__init__(name)
        self.name = name

# Python 文件
``` python
# 打开文件
f = open("///", encoding = "utf-8")
print(f.read()) # 阅读全部内容
print(f.readline()) # 只读取一行

line = f.readline()
while line != "":
    print(line)
    line = f.readline()

print(f.readlines()) # 读取指定行数

f.close()

# 自动关闭
with open("///") as f:
    #///

```

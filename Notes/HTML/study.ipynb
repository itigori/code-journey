# HTML基本骨架

```HTML
<html>
    <head>
        <title>网页标题</title>
    </head>
    <body>
        网页主体
    </body>
<html>
```

html：整个网页
head：网页头部，存放浏览器代码，比如css
body：网页主体，存放给用户看的代码，比如图片，文字
title：网页标题

vs code中快速生成代码
! + Enter/Tab

有序列表：排版有规定顺序的列表
```html
<!-- 父级 -->
<ol> 
    <!-- 子级 -->
    <li></li>
</ol>

<!-- ol只能包含li -->
```

定义列表：一个标题对应一个列表
```html
<!-- 父级 -->
<dl> 
    <!-- 子级 -->
    <dt>标题</dt>
    <dd>描述</dd>
</dl>
```

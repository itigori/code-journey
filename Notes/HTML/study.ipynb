# HTML定义
超文本标记语言

- 标签成对出现
- 结束标签多一个/
- 包裹内容是双标签，其他是单标签

# HTML基本骨架

```HTML
<html>
    <head>
        <title>网页标题</title>
    </head>
    <body>
        网页主体
    </body>
<html>
```

html：整个网页
head：网页头部，存放浏览器代码，比如css
body：网页主体，存放给用户看的代码，比如图片，文字
title：网页标题

vs code中快速生成代码

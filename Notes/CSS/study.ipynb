# CSS

Cascading style sheets 层叠样式表

书写方式：title标签下方添加style标签

```HTML
<html>
    <head>
        <title>体验CSS</title>
        <style>
            /* CSS 代码 */
            /* 选择器 {} */
            /* 键值对 */
            p {
                /* color：颜色，font-size：大小 */
                color: #66CCFF;
                font-size: 30px;
            }
        </style>
    </head>
    <body>
        <P>体验CSS</P>
    </body>
</html>
```

# CSS 引入方式

- 内部样式表：写在style标签内

- 外部样式表：开发使用 => 放在css文件中 => HTML中使用Link连接

``` HTML
<link rel="stylesheet" href="./my.css">
```

- 行内样式：配合JavaScript使用
```HTML
<div style="color:red;font-size:30px;"></div>
```


# 选择器

查找标签，设置样式

- 标签选择器都以标签命名，选中相同标签设置相同属性

- 类选择器，差异化同名标签 => 定义类名.类名 => 使用class=""

- id选择器，查找标签设置不同样式 => 定义id选择器#id => 使用id=""（同一个id在同一页面只能使用一次）

- 通配符选择器，查找页面所有标签，设置相同样式 => *

``` CSS
* {
    color :red;
}
```
